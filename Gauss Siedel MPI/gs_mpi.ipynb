```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'D:\Masters\Spring2023\CprE 525 Numerical Analysis for HPC\example4\data.csv', sep=',')
```

```python
data.head()
```

```python
data = data.drop('Unnamed: 3', axis=1)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(data)
```

```python
grouped_data = data.groupby('Nodes')['Total-elapsed'].sum()

# create bar plot
plt.bar(grouped_data.index, grouped_data.values)

# add labels and title
plt.xlabel('Nodes')
plt.ylabel('Total-elapsed')
plt.title('Total Elapsed Time vs Nodes')

# show plot
plt.show()
```

```python
# group data by Nodes
grouped_data = data.groupby('Nodes')

# create line plot for each group
for name, group in grouped_data:
    plt.plot(group['Total-elapsed'], group['Size'], label=f'Nodes={name}')

# add labels and title
plt.xlabel('Total-elapsed')
plt.ylabel('Size')
plt.title('Size vs Time-elapsed for Each Value of Nodes')

# add legend
plt.legend()

# show plot
plt.show()
```

```python
# group data by Nodes
grouped_data = data.groupby('Nodes')

# create line plot for each group
for name, group in grouped_data:
    plt.plot(group['Total-elapsed'], group['Nodes'], label=f'Size={name}')

# add labels and title
plt.xlabel('Total-elapsed')
plt.ylabel('Size')
plt.title('Size vs Time-elapsed for Each Value of Nodes')

# add legend
plt.legend()

# show plot
plt.show()
```

```python
# create a pivot table with Size as the rows, Nodes as the columns, and Total-elapsed as the values
pivot = data.pivot(index='Size', columns='Nodes', values='Total-elapsed')

# create the heatmap
sns.heatmap(pivot, cmap='YlOrRd')

# add labels and title
plt.xlabel('Nodes')
plt.ylabel('Size')
plt.title('Size vs Nodes with Total-elapsed')
plt.show()
```

```python
# create the boxplot
sns.boxplot(x='Nodes', y='Total-elapsed', data=data)

# add labels and title
plt.xlabel('Nodes')
plt.ylabel('Total-elapsed')
plt.title('Total-elapsed for each value of Nodes')
plt.show()
```
